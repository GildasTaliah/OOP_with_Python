# Intro
Object Oriented Programming (OOP) is a style of programming that heavily relies on objects. Objects are instances of classes, that encapsulates data and behaviour, put it diferrently they have attributes and methods. Attribute is said to store data while methods defines behaviour.

Like any other programming language Python supports OOP, functional programming, decorators etc. However, OOP becomes much more valuable when writing large-sized and complex programs that demand a strucured and modular approach to manage complexity, promote code organization and fascilitate collaboration amongt team members.

Below I present my first Python class, that takes in a number (integer, real number...) and does a bunch of computations with it: adds another number to it, doubles the number, multiply number by an input - just a bunch of random stuff in general.


```python
# Create class
class ManipulateNumber:

  # Create an instance
  def __init__ (my_first, x , y = 2):
    my_first.x = x           # attribute
    my_first.y = y           # attribute with default value

  ## Methods
  # Multiply x by n
  def multi (my_first, n = None):
    if n == None:
      return my_first.x * my_first.y
    else:
      return my_first.x * n

  # Double x
  def double(my_first):
    return my_first.x * my_first.y

  # Add n to x
  def add (my_first, n = None):
    if n == None:
      return my_first.x + my_first.y
    else:
      return my_first.x + n

  # Divide x by n
  def div(my_first, n = None):
    if n == None :
      return my_first.x / my_first.y
    else:
      return my_first.x / n
```

```python

fivehund = ManipulateNumber(500)
print(f' Doubling {500} results to: {fivehund.double()}')
print(f' Multiplying {500} by {0.95} yields: {fivehund.multi(.95)}')
print(f' Adding {-50} to {500} returns: {fivehund.add(-50)}')
print(f' Dividing {500} by {1.25} results to: {fivehund.div(1.25)}')
```

## The Vector Class

Our next example is to define a Vector class. A vector is an array of elements, representing quantities in both magnitude and direction. It is different from a matrix in that a matrix is multi-dimentional. The Vector class below, has both class and instance attributes. The instance attribute has default parameters, and the class has several defined methods to perform different operations.


```python
# Define a Vector Class that inherits from default/built-in object
class Vector(object):
  # Define the class attribute
  info = 'Does a bunch of operations with the input vector. Default values: 0, 0 & 0.'

  # Define an instnace with attributes: x, y, z
  def __init__(self, x=0, y=0, z=0):
    self.x = x # Age of young childdren
    self.y = y # Height
    self.z = z # Weights

  # Define methods
  def __repr__ (self):
    return 'Vector (%r, %r, %r)' % (self.x, self.y, self.z)

  def __abs__(self):
    return (self.x ** 2 + self.y ** 2 + self.z ** 2) ** 0.5

  def __bool__(self):
    return bool(abs(self))

  ## Vector Oprations
  # Add new vector to Vector
  # (-new vector) To subtract new vector from Vector
  def __add__(self, new_vec):
    x = self.x + new_vec.x
    y = self.y + new_vec.y
    z = self.z + new_vec.z
    return Vector(x, y, z)

  # Scalar multiplication
  def __mul__(self, scalar):
    return Vector(self.x * scalar,
                  self.y * scalar,
                  self.z * scalar)

  # Dot product
  def __dotproduct__(self, new_vec):
    a, b, c = self.x * new_vec[0], self.y * new_vec[1], self.x * new_vec[2]
    return sum([a, b, c])

  # Cross product
  def __crossproduct__(self, new_vec):
    a, b, c = self.x * new_vec[0], self.y * new_vec[1], self.x * new_vec[2]
    return Vector(a, b, c)

  # Vector lenght
  def __len__(self):
    return 3

  # Get vector item.
  def __getitem__(self, i):
    if i in [17, 27]: return 'Age: ' + str(self.x)
    elif i in [150, 195]: return 'Height: ' + str(self.y)
    elif i in [65, 120]: return 'Weight: ' + str(self.z)
    else: raise IndexError('Index out of range.')

  # Iterate over vector items
  def __iter__(self):
    for i in  range(len(self)):
      yield self[i]
```

```python
# Create an instance: input a vector
cvec = list([27, 175, 65])
vec = Vector(*cvec) # unpack cvec
vec.info
```

```python
# The - represent method
print(f' The represent method: {vec.__repr__()}')

# The - absolute value method
print(f' The absolute value method: {vec.__abs__()}')

# The - bool method
print(f' The bool method: {vec.__bool__()}')
```

```python
# @title vector operations
# The - add method
print(f' Add {[-2, 10, 5]} to {vec} yields: {vec.__add__(Vector(x=-2, y=-10, z=-5))}')

# Scalar multiplication
vec.__mul__(1.05)
print(f' Multiply {vec } by {1.05} results to : {vec.__mul__(1.05)}')

# Dot product
vec.__mul__(1.05)
print(f' The dotproduct between  {vec} and {[10, 20, 30]} returns: {vec.__dotproduct__([10, 20, 30])}')

# Cross product
print(f' The Cross product between {vec} and {[10, 20, 30]} results to : {vec.__crossproduct__([10, 20, 30])}')
```

```python
# The - lenght method
print(f'The lenght of {vec} is: {vec.__len__()}')

# The - get item method
print(vec.__getitem__(i=27))

# Generator object
print(vec.__iter__())
```

## Encapsulation

Encapsulation a core concept of OOP, refers to the bundling of data (attributes) and methods within a class. It provides data protection and control over how the codes interact with an object internal state. Encapsulation can be achieved by defining private attributes and methods within a class. By convention, private attributes and methods are prefixed with a single underscore (_).

We define the HybridHVAC (Heating, ventilation and Air Conditioning) class below to illustrate this new concept. Additionally encapsulation promotes code modularity, maintainability, and data protection by separating the internal state from the public interface (methods). It allows us to change the internal state without affecting the code that uses the class, as long as the public interface remains the same.






```python

# we define a class - HybridHVAC
# Also define methods - get_colour() and _set_colour() acts as interface
# for interacting with the internal state of the HybridHVAC class
class HybridHVAC:

    Brand = "Lucky Goldstar - LG"

    def __init__(self, colour, model):
        self._colour = colour  # Private attribute
        self._model = model    # Private attribute

    def Hpower_on(self):
        print(f'Powering on {self._colour} {self._model} Heater.')

    def Hpower_off(self):
        print(f'Powering off {self._colour} {self._model} Heater.')

    def Cpower_on(self):
        print(f'Powering on {self._colour} {self._model} Cooler.')

    def Cpower_off(self):
        print(f'Powering off {self._colour} {self._model} Cooler.')

    def get_colour(self):
        return self._colour

    def _set_colour(self, new_colour): # Private method
        self._colour = new_colour


heater_one = HybridHVAC("black", "A8eWd")
heater_one.Hpower_on()

cooler_one = HybridHVAC("whhtt", "D5vQh")
cooler_one.Cpower_on()

# Accessing a private attribute directly (not advised)
print(cooler_one._colour)

# Accessing a private attribute using the getter method (preferred)
print(cooler_one.get_colour())
print(heater_one.get_colour())


# Modifying a private attribute using the setter method (preferred)
cooler_one._set_colour("white")
print('*' * 3, 'After updating Cooler one colour to white', '*' * 3)
print(cooler_one.get_colour() )
print(heater_one.get_colour())
```

# Black Scholes Merton Model

A derivative is a financial instrument whose price depends on the performance of an underlying asset. Underlyings can be, for example, stocks, bonds, interest rates, commodities or foreign currencies.

Derivatives play a very crucial role in modern financial markets by providing tools for risk management/hedging, speculation, arbitrage and investment strategies.  Examples of derivatives include options, futures, forwards and swaps.

In here, our focus will be on pricing European options, an option contract gives the holder the right, but not the obligation to buy (call option) or sell (put option) a fixed quantity of an underlying asset at a predefined price, within a specified timeframe. For an agreement to be made a fair value (or fair price) needs to have been derived.   

The single most important option pricing model is probably the Black-Scholes-Merton Model *BSM*. It is a mathematical model used to calculate the theoretical price of European-style options, and used as benchmark for other option pricing models such as Monte Carlo (MC), binomial model, integral transform, and finite differences.

The model takes into account factors such as the underlying current stock price $S$, the option's strike price $K$, the time until maturity $T$, the risk-free interest rate $r$, and the volatility of the underlying asset $σ$. The option price of a call (C) and put (P) option is computed as follows:


$$C = S \cdot N(d_1) - K \cdot e^{-rT} \cdot N(d_2)\$$


$$P = K \cdot e^{-rT} \cdot N(-d_2) - S \cdot N(-d_1)
$$


Where $N(d_1)$ and $N(d_2)$  are the cumulative standard normal distribution functions of *d1* and *d2*, and are computed as follows:

$$d_1 = \frac{\ln\left(\frac{S}{K}\right) + \left(r + \frac{{\sigma^2}}{2}\right)T}{\sigma \sqrt{T}}
$$

$$ d_2 = d_1 - \sigma \sqrt{T}$$.

For comparison we will compute the option price using the monte carlo technique. The MC estimator for the european call option price is presented as follows:

$$ C = \exp^{-r T} \frac{1}{M} \sum_{m=1}^{M} \max \left((S_{T}(m) - K), 0\right)$$

M is the number of MC simulated values. In addition to the above we will compute the delta and vega of an option. Delta measures the sensitivity of the option price with respect to the changes in the underlying price, and vega measure the sensitivity of the option price to changes in volatility (or implied volatility). Lastly we develop a method to compute implied volatility for quoted option price.

## Assumptions of the Black Scoles Model

While we have understood how to price an option using the Black Scholes Merton model, and before we dive into the python implementation, it is imperative to understand the assumptions behind the model. We outlined these asuumptions below:

- There are no arbitrage, that is no possibility of making riskless profits.
- It is possible to borrow and lend cash at a risk-free rate, and the market is frictionless (no transaction cost).
- There is no restraint on short-selling stocks, and the stock prices follow a Geometric Brownian Motion.

The model also assumes no dividend payment. However, under the assumption that that both the dividend and the risk-free rate are continuously compounding, dividends can be taken into account by subtracting them from the risk-free rate. Nonetheless,  It is also worth noting that some assumptions may not hold in practical situations or at least with some assets (classes). For example with less liquid asset the frictionless market assumption would be violated.

```python
# Import necessary packages
import numpy as np
import pandas as pd
from scipy import stats
```

```python

class bsm_option:

  ''' Class for European call and put options price computed under the Black
  Scholes Merton model
  and MonteCarlo Simulations.

  Attributes
  ==========
  S0, K, T, r, sigma.

  Methods
  =======
  value_call, value_put, mc_sim, delta_call, delta_put, vega and
  implied volatility

  '''


  def __init__(self, S0, K, T, r, sigma):
    self.S0 = float(S0)    # Initial Value
    self.K = K             # Strike at maturity
    self.T = T             # Maturity in years
    self.r = r             # risk free short rate
    self.sigma = sigma     # volatility

    # Set delta and vega values initially to zero
    self.delta = 0
    self.veg = 0


  def value_call(self):
    ''' Returns call option value and updates the delta value after computing d1
    '''
    d1 = ((np.log(self.S0/self.K) + (self.r + 0.5 * self.sigma ** 2) * self.T) /
         (self.sigma * np.sqrt(self.T)))
    # update delta value
    self.delta = d1

    #d2 = ((np.log(self.S0/self.K) + (self.r - 0.5 * self.sigma ** 2) * self.T) /
      #    (self.sigma * np.sqrt(self.T)))
    d2 = d1 - self.sigma * np.sqrt(self.T)

    value = (self.S0 * stats.norm.cdf(d1, 0.0, 1.0)  -
            self.K * np.exp(-self.r * self.T) * stats.norm.cdf(d2, 0.0, 1.0))
    return value

  def value_put(self):
    ''' Returns put option value
    '''
    d1 = ((np.log(self.S0/self.K) + (self.r + 0.5 * self.sigma ** 2) * self.T) /
        (self.sigma * np.sqrt(self.T)))

    # This is same as d2 above, just another way around
    d2 = d1 - self.sigma * np.sqrt(self.T)

    value = (self.K * np.exp(-self.r * self.T) * stats.norm.cdf(-d2, 0.0, 1.0) -
          self.S0 * stats.norm.cdf(-d1, 0.0, 1.0))
    return value

  def mc_sim(self, option = 'call', MC_Sim = 100_000):
    ''' Computes the Monte Carlo option price for call and put options with
      100_000 simulations.

      - Computes call by default or explicitly when writen call.
      - Computes put when 'put' explicitly written
    '''
    # Standard normally distributed numbers: 0 mean, unit variance.
    z = np.random.standard_normal(MC_Sim)

      # Values at maturity
    ST = (self.S0 * np.exp((self.r - 0.5 * self.sigma ** 2) * self.T  +
            self.sigma * np.sqrt(self.T) * z))

      # payoff at maturity
    if option == 'call' or None:
      CPT = np.maximum(ST - self.K, 0)
      return np.exp(-self.r  *  self.T) * np.mean(CPT)
    elif option == 'put':
      PPT = np.maximum(self.K - ST, 0)
      return np.exp(-self.r * self.T) * np.mean(PPT)
    else:
      print("Invalid entry. Input either 'call' or 'put'. ")


  def delta_call(self):
    ''' Returns delta. Delta measures the sensitivity of the option price to changes
    in the underlying asset price.
    - the first partial derivative of the option price with respect to asset price
    '''
    # return delta for call option
    d1 = ((np.log(self.S0/self.K) + (self.r + 0.5 * self.sigma ** 2) * self.T) /
        (self.sigma * np.sqrt(self.T)))


    return self.delta #, d1

  def delta_put(self):

    # return delta for put option
    return self.delta - 1

  def vega(self):
    ''' Returns vega. Vega measures the sensitivity of the option price to
    changes in the implied volatility of the underlying price.
    - Will be used to compute implied volatility
    '''

    d1 = ((np.log(self.S0/self.K) + (self.r + 0.5 * self.sigma ** 2) * self.T) /
        (self.sigma * np.sqrt(self.T)))
    vega = self.S0  * np.sqrt(self.T) * stats.norm.pdf(d1, 0.0, 1.0)
    self.veg = vega
    return vega

  def implied_vol(self, C0, it = 100):
    ''' Returns the implied volatility given the obseerved (quoted) call option price
    '''
    option = bsm_option(self.S0, self.K, self.T, self.r, self.sigma)
    for i in range(it):
      option.sigma -= (option.value_call() - C0) / option.vega()
    return option.sigma
```

```python
# An option can be 'in the money', 'out of the money' or 'at the money'.
# Given Stock price: 100 €, Strike = 105 €, Maturity(T) = 1 year,
# risk-free interest rate = 5 %, stock volatility = 20 %
# Compute option price
option = bsm_option(100, 105, 1.0, 0.05, 0.2)

print('-' * 5, 'OUT OF THE MONEY call option' , 5 * '-')
print(f' Call option price BSM: {option.value_call()}')
print(f' Call option price MC: {option.mc_sim()}')
print('-' * 20 , '****', 20 * '-')

print('-' * 5 , 'IN THE MONEY put option' , 10 * '-')
print(f' Put option price BSM: {option.value_put()}')
print(f" Put option price MC: {option.mc_sim('put')}")
print('-' * 20 , '****', 20 * '-')

print(f' Delta call: {option.delta_call()}')
print(f' Delta put: {option.delta_put()}')
print('-' * 20 , '****', 20 * '-')

#print(f' Vega of option price: {option.vega()}')
print(f' Implied Vol with option value quoted at 15€: {option.implied_vol(15)}')
```

## Sensitivity Analysis

One of the major benefits of an analytical option pricing formula is that it can be used to visualize for instance the option value, the vega or delta of the option for different strikes and maturities. We proceed to generate these statistics for different maturity-strike combinations.

```python
# create an instance
o = bsm_option(105, 100, 1, 0.05, 0.2)

# create range of K and T
# loop through and obtain option and vega values
# prepare data for plotting
maturities = np.linspace(0.01, 1.5, 20)
strikes = np.linspace(85, 120, 20)
K, T = np.meshgrid(strikes, maturities)

C = np.zeros_like(K)
V = np.zeros_like(C)

for t in enumerate(maturities):
  for k in enumerate(strikes):

    o.T = t[1]         # update maturity value
    o.K = k[1]         # update strike value
    # compute new call option value and vega with updated t and k values
    C[t[0], k[0]] = o.value_call()
    V[t[0], k[0]] = o.vega()
```

```python
from pylab import cm, mpl, plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline


fig = plt.figure(figsize = (15, 7.5))
#ax = fig.gca()

# Add a 3D subplot to the figure
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(K, T, C, rstride = 1, cstride = 1, cmap = cm.coolwarm,
                       linewidth= 0.5, antialiased = True)

ax.set_xlabel('strike - K')
ax.set_ylabel('maturity - T')
ax.set_zlabel('Value European Call')
fig.colorbar(surf, shrink = 0.45, aspect = 7);
```

The value per option contract increases towards 25 euro, for in the money call options (strike less than current price of 105€) as the exercise period is further away.

```python
fig = plt.figure(figsize = (15, 8))
#ax = fig.gca()

# Add a 3D subplot to the figure
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(K, T, V, rstride = 1, cstride = 1, cmap = cm.coolwarm,
                       linewidth= 0.5, antialiased = True)

ax.set_xlabel('strike - K')
ax.set_ylabel('maturity - T')
ax.set_zlabel('vega')
fig.colorbar(surf, shrink = 0.45, aspect = 6);
```

Option price sensitivity to volatility (vega) increases towards 50, for higher strikes (out of the money call)  with expiration period further away.
